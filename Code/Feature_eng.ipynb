```python
y = df['label']
df.drop(columns=['label'], inplace=True,axis= 1)
```

```python
numerical = df.select_dtypes(include = 'number')
cat = df.select_dtypes(exclude = 'number')
```

```python
# robust scaler
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
df[numerical.columns] = rs.fit_transform(numerical)
```

```python
df[numerical.columns].head()
```

```python
# mutual info score
from sklearn.metrics import mutual_info_score
import pandas as pd

# One-hot encode the categorical features
df_encoded = pd.get_dummies(df, columns=cat.columns, drop_first=True)

# Calculate mutual info scores for each feature
mi_scores = df_encoded.apply(lambda x: mutual_info_score(y, x))

print(mi_scores)
```

```python
# split into train and valid
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(df_encoded, y, test_size=0.2, random_state=42)
```

```python
# simple models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report,auc,roc_auc_score,roc_curve
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'AdaBoost': AdaBoostClassifier(),
    'XGBoost': XGBClassifier()
}
```

```python
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    class_report = classification_report(y_valid, y_pred)
    print(f"{name} Classification Report:\n{class_report}\n")
    print(f"{name} Confusion Matrix:\n{confusion_matrix(y_valid, y_pred)}\n")
    roc_auc = roc_auc_score(y_valid, y_pred)
    print(f"{name} ROC AUC Score: {roc_auc}\n")
    results[name] = {
        'accuracy': accuracy_score(y_valid, y_pred),
        'precision': precision_score(y_valid, y_pred),
        'recall': recall_score(y_valid, y_pred),
        'f1_score': f1_score(y_valid, y_pred),
        'roc_auc': roc_auc,
    }
```

```python
# get y_proba
for name, model in models.items():
    y_prob = model.predict_proba(X_valid)[:, 1]
    results[name]['y_prob'] = y_prob
```

```python
plt.figure(figsize=(10, 7))
for name, res in results.items():
    fpr, tpr, _ = roc_curve(y_valid, res['y_prob'])
    plt.plot(fpr, tpr, label=f"{name} (AUC = {res['roc_auc']:.2f})")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.grid()
plt.show()
```

```python
# feature importance for xgboost
importances = models['XGBoost'].feature_importances_
indices = np.argsort(importances)[::-1]
imp_df = pd.DataFrame({'Feature': df_encoded.columns[indices], 'Importance': importances[indices]})
imp_df.head(20)
```

```python
# plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=imp_df.head(10))
plt.title('Top 10 Feature Importances')
plt.show()
```

```python
# select features bigger than threshold
def select_features(model,x_train, x_valid, x_test, y_train, threshold):
    importances = model.feature_importances_
    indices = np.where(importances > threshold)
    selected_features = x_train.columns[indices]
    x_train_selected = x_train[selected_features]
    x_valid_selected = x_valid[selected_features]
    x_test_selected = x_test[selected_features]
    return x_train_selected, x_valid_selected, x_test_selected
```

```python
def train_and_evaluate_model(model, x_train, y_train, x_valid, y_valid):
    model.fit(x_train, y_train)
    y_pred_valid = model.predict(x_valid)
    print(classification_report(y_valid, y_pred_valid))
    print(confusion_matrix(y_valid, y_pred_valid))
    print(roc_auc_score(y_valid, y_pred_valid))
```

```python
# try diffrent thresholds
thresholds = [0.001,0.01,0.1,0.2]
for threshold in thresholds:
    x_train_selected, x_valid_selected, x_test_selected = select_features(X_train, X_valid, x_test, y_train, threshold)
    print(f"Threshold: {threshold}")
    xgb = XGBClassifier()
    train_and_evaluate_model(xgb, x_train_selected, y_train, x_valid_selected, y_valid, x_test_selected, y_test)
```
